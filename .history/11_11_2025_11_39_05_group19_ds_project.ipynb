5. What the variables mean
a. players: variables in character type and logical type are qualitative information, and the variables double types are quantitative information, but the variable in logical type could be recognized as categorical type.
- experience: how experienced the players are in the games by classifying them into different levels: “Amateur”, “Beginner”, “Pro”, “Regular”, “Veteran”.
- subscribe: whether players subscribe to a game-related newsletter, indicated with TRUE and FALSE
- hashedEmail: players’ email address
- played_hours: players’ total playing durations in the server in hours
- name: name of players
- gender: gender of players
- Age: age of players

- In sessions, hashEmail is qualitative information, which are the emails of the players in the recorded session; there are start time and end time in character type indicated the exact date and time of the sessions start and end with human-readable format, original start time and original end time are in double type that are not readable by users.
- hashedEmail: players’ email address
- start_time: start time of each play session in **date format** (dd/mm/yyyy & time)
- end_time: end time of each play session in **date format** (dd/mm/yyyy & time)
- original_start_time: start time of each play session in **seconds**
- original_end_time: end time of each play session in **seconds**
